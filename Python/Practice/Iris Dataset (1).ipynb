### Logistic Regression

Even though Logistic Regression contains the word 'Regression', it actually performs classification. The term 'Logistic' comes from the Logistic Function (also known as the sigmoid function).

Sigmoid function f(x) = ${\large \frac{1}{1 + e^{-x}}}$

The range of sogmoid function is (0, 1).

Equation of Logistic Regression: $\hat{y} = {\large\frac{1}{1+e^{-(\beta_0+\Sigma\beta_i.X_i)}}}$

The derivation of cost function and also the method for calculation is almost same as that of linear regression.

#### Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression(max_iter=200)
```

```python
log_model.fit(X_train, y_train)
```

```python
log_pred = log_model.predict(X_test)
```

```python
log_pred[:10]
```

```python
y_test[:10]
```

### Performance Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report
```

#### Confusion Matrix

**Rows:** Actually Setosa, Actually Versicolor, Actually Virginica  
**Columns:** Predicted Setosa, Predicted Versicolor, Predicted Virginica

```python
print(confusion_matrix(y_test, log_pred))
```

```python
pd.DataFrame(
    data=confusion_matrix(y_test, log_pred),
    index=['Actually Setosa', 'Actually Versicolor', 'Actually Virginica'],
    columns=['Predicted Setosa', 'Predicted Versicolor', 'Predicted Virginica']
)
```

#### Classification Report

```python
print(classification_report(y_test, log_pred))
```

**Accuracy:** Percentage of predictions that are correct.

**Precision:** True Positive / Total Predicted Positive (Ability of the model to find all relevant data points)

**Recall:** True Positive / Total Actual Positive (Ability of model to find all relevant cases)

**F1-Score:** ${\large\frac{1}{f1-score}=\frac{1}{2}(\frac{1}{precision}+\frac{1}{recall})}$ (Harmonic mean of precision and recall)

**Write a function that will take a list of models and will return a list of accuracy scores corresponding to each model.**

```python
from sklearn.metrics import accuracy_score

def func(models):
    res = []
    for i in range(len(models)):
        models[i].fit(X_train,y_train)
        res_pred = models[i].predict(X_test)
        res.append(accuracy_score(y_test,res_pred))
    return res
```

```python
models = [
    LogisticRegression(max_iter=200),
    DecisionTreeClassifier(),
    RandomForestClassifier(n_estimators=1000)
]
print(func(models))
```

### K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors=5)
```

```python
knn_model.fit(X_train, y_train)
```

```python
knn_pred = knn_model.predict(X_test)
```

```python
print(confusion_matrix(y_test, knn_pred))
```

```python
print(classification_report(y_test, knn_pred))
```

```python
error = []
for i in range(1, 31):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    error.append(np.mean(pred != y_test))
```

```python
plt.plot(range(1, 31), error, color='k', ls='dashed', marker='o', markerfacecolor='r')
plt.title('Error Rate vs. n_neighbors(k)')
plt.xlabel('n_neighbors')
plt.ylabel('Error Rate')
plt.show()
```

```python
knn_model = KNeighborsClassifier(n_neighbors=7)
knn_model.fit(X_train, y_train)
knn_pred = knn_model.predict(X_test)
print(confusion_matrix(y_test, knn_pred))
print(classification_report(y_test, knn_pred))
```
